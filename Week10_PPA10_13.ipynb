```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)

z = np.cos(x)
plt.plot(x,z)
```

```python
f = open("scores.csv",'r')
marks = []
x_val = []
y_val = []
f.readline()
line = f.readline()
while(line != ''):
  marks = line.strip().split(',')
  x_val.append(marks[0])
  y_val.append(marks[1])
  line = f.readline()
#print(x_val,y_val)
plt.scatter(x_val,y_val)
```

```python
import random
n = 1000000
c1 = 0;c2=0;c3=0;c4=0;c5=0;c6=0
for i in range(n):
  f = random.randint(1,6)
  if(f==1):
    c1 += 1
  if(f==2):
    c2 += 1
  if(f==3):
    c3 += 1
  if(f==4):
    c4 += 1
  if(f==5):
    c5 += 1
  if(f==6):
    c6 += 1
p1 = c1/n
p2 = c2/n
p3 = c3/n
p4 = c4/n
p5 = c5/n
p6 = c6/n
p = [p1,p2,p3,p4,p5,p6]
x = [1,2,3,4,5,6]
print(p)
plt.bar(x,p)
```
