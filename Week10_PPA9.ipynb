```python
import numpy as np
import matplotlib.pyplot as plt

x_val = np.linspace(0,10,50)
y_val = (3* x_val) + 4
plt.plot(x_val,y_val)
```

```python
x_val = np.linspace(-10,10,50)
y_val = (x_val**2)+(2*x_val)-15
plt.plot(x_val,y_val)
```

```python
x_val = np.linspace(-100,100,100)
y_val = 5*(x_val-1)*(x_val-2)*(x_val-3)
plt.plot(x_val,y_val)
```

```python
x_val = np.linspace(0,5,50)
y_val = np.exp(x_val)

plt.plot(x_val,y_val)
```

```python
x_val = np.linspace(1,10,100)
y_val = np.log(x_val)
plt.plot(x_val,y_val)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)
```
